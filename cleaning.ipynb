```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
import numpy as np
nltk.download('words')
en_words = set(nltk.corpus.words.words())
regex = re.compile('[^a-zA-Z]')
nltk.download('stopwords')
porter = PorterStemmer()
```

```python
def clean_app_desc(desc):
    lower_word = regex.sub(' ',str(desc).lower())
    cleaned_uri = re.sub(r'http\S+', '', lower_word)
    stemming_word = porter.stem(cleaned_uri)
    punc_eng_word = [w for w in nltk.wordpunct_tokenize(stemming_word) if w.lower() in en_words or not w.isalpha()]
    cleaned_stopword = [word for word in punc_eng_word if word not in stopwords.words('english')]
    return ((' '.join(cleaned_stopword)).replace('\n','')).replace('\t','')
```
